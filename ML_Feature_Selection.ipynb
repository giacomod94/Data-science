```python
#define a function to compare MAE from different startng points
def score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid):
    model = RandomForestRegressor(random_state=1)
    model.fit(reduced_X_train, y_train)
    preds_val = model.predict(reduced_X_valid)
    mae = mean_absolute_error(y_valid, preds_val)
    return(mae)
```

```python

def learningCurve(my_pipeline,X_train,y_train,X_valid,y_valid,n):
    valid = []
    train = []
    k = 1
    lt = math.ceil(len(X_train)/n)
    
    for m in range(n):
        i = lt*k
        
        # Preprocessing of training data, fit model 
        my_pipeline.fit(X_train[0:i], y_train[0:i])

        # Preprocessing of validation data, get predictions
        preds_valid = my_pipeline.predict(X_valid)
        preds_train = my_pipeline.predict(X_train[0:i])

        # Evaluate the model
        valid.append(mean_absolute_error(y_valid, preds_valid))
        train.append(mean_absolute_error(y_train[0:i], preds_train))
        k = k+1
        
    print(valid)
    print(train)
    x = np.arange(lt+len(X_valid),len(X_train)+len(X_valid)+lt,lt)
    plt.plot(x,valid)
    plt.plot(x,train)
    plt.xlabel('Data', fontsize = 10)
    plt.ylabel('MAE', fontsize = 10)
    plt.title('LearningCurve', fontsize = 15)
    plt.show()
    
        
        
        
```
