```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\Giacomo\Desktop\polimi\tesi\SIEMENS\Dataturbine1ok.csv')
data['MeasuringTime'] = pd.to_datetime(data['MeasuringTime'], format= str(['%Y-%m-%d %H:%M:%S'])) #turbine1
#data['MeasuringTime'] = pd.to_datetime(data2['MeasuringTime'], format= str(['%d/%m/%Y %H:%M'])) #turbine2,3,4
data = data.drop('Unnamed: 0', axis = 'columns')
data = data.set_index('MeasuringTime');# set MeasuringTime as index
data.head()
```

```python
#histograms
data.hist(figsize = (15,10))
plt.show()
```

```python
#histogram
data.actpower_mean.hist(figsize = (15,10))
plt.xlabel('points', fontsize = 25)
plt.ylabel('values', fontsize = 25)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15, rotation = 45)
plt.show()
```

```python
#scatterplot

ax = plt.figure(figsize = (15,10))
sns.scatterplot(x = 'winavail_mean', y = 'actpower_mean', data = data, hue = 'winavail_SD')

plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15)
plt.show()
```

```python
#Crosscorrelaztion

f = plt.figure(figsize=(19, 15))
plt.matshow(data.drop('turbinok', axis = 'columns').corr(), fignum=f.number, )

#add features name
plt.xticks(range(data.drop('turbinok', axis = 'columns').shape[1]), data.drop('turbinok', axis = 'columns').columns, fontsize=14, rotation=45)
plt.yticks(range(data.drop('turbinok', axis = 'columns').shape[1]), data.drop('turbinok', axis = 'columns').columns, fontsize=14)

#adda colorbar
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)

plt.autoscale()
plt.show()

```
